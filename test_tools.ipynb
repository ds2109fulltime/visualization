## Testing functions

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from visualization_funcs import draw_feat_importance, missing_ratio, statistic_values, target_transform

from sklearn.linear_model import LinearRegression
```

### Test: draw_feat_importance

```python
diamonds_df = sns.load_dataset('diamonds')

diamonds_df2 = pd.get_dummies(diamonds_df)

lr = LinearRegression()
X = diamonds_df2.drop(columns = ['price'])
y = diamonds_df2['price']

lr.fit(X, y)

draw_feat_importance(lr.coef_, diamonds_df2.drop(columns = ['price']).columns, 'Testing', figsize=(15,8))
```
